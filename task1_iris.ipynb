# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction
The Iris dataset is one of the most well-known datasets in data science and machine learning.
It contains 150 samples of iris flowers across 3 species:
- Iris Setosa
- Iris Versicolor
- Iris Virginica

Each sample has 4 features: sepal length, sepal width, petal length, and petal width (all in cm).

## Objective
Understand how to load, summarize, and visualize a dataset using pandas, matplotlib, and seaborn.

```python
# Import all required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for all plots
sns.set_theme(style="whitegrid")
print("Libraries imported successfully!")
```

```python
# Load the Iris dataset from seaborn's built-in datasets
df = sns.load_dataset('iris')

print("✅ Dataset loaded successfully!")
print(f"Shape (rows, columns): {df.shape}")
```

```python
# Display the first 5 rows
print("First 5 rows:")
df.head()
```

```python
# Display column names and data types
print("Column Names:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nBasic Statistics:")
df.describe()
```

```python
# Check for any missing/null values in the dataset
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
# How many samples per species?
print("Sample count per species:")
print(df['species'].value_counts())
```

```python
# Scatter plot: Sepal Length vs Sepal Width, colored by species
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set2', s=80)
plt.title('Sepal Length vs Sepal Width by Species', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Histogram: distribution of all 4 features
df.drop(columns='species').hist(figsize=(10, 6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distributions - Iris Dataset', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Box plot: spread and outliers for each feature by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(data=df, x='species', y=feature, ax=ax)
    ax.set_title(f'{feature.replace("_", " ").title()} by Species')
    ax.set_xlabel('Species')

plt.suptitle('Box Plots: Feature Spread by Species', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Pairplot: relationships between all features at once
sns.pairplot(df, hue='species', palette='Set2', diag_kind='kde')
plt.suptitle('Pairplot of Iris Features', y=1.02, fontsize=14)
plt.show()
```

## Conclusion

From the Iris dataset exploration, we found:

- **No missing values** — the dataset is clean and ready to use.
- **Setosa** is clearly separable from the other two species based on petal length and petal width.
- **Versicolor and Virginica** overlap slightly, making them harder to distinguish.
- **Petal features** are stronger indicators of species than sepal features.
- The dataset is perfectly **balanced** — 50 samples per species.

This EDA gives a strong foundation for building a classification model in future tasks.
